```r
x  <- mpg$hwy
y <- seq(1,234)
```

```r
df <- data.frame(x,y)
distxy <- dist(df)
hc <- hclust(distxy)
plot(hc)
abline(h=100, col="red", lwd=3)
```

```r
heatmap(as.matrix(df))
```

```r
kmeansObj <- kmeans(df,centers=4)
plot(df,col=kmeansObj$cluster)
points(kmeansObj$clusters, pch=3)
```

```r
set.seed(111)
x <- rnorm(30,mean = rep(1:3,each=3), sd=0.3)
y <- rnorm(30,mean = rep(1:3, each=3), sd=0.5)
df <- data.frame(x,y)
kmeansObj <- kmeans(df, centers=3)
plot(df,col=kmeansObj$cluster)
points(kmeansObj$centers, pch=3, cex=3,col=1:3,lwd=4)
```

```r
image(t(df))
```

```r
par(mfrow=c(2,2))
plot(df,main="nothing")
plot(svd(scale(df))$u[,1],main="u")
plot(svd(scale(df))$d,main="d")
plot(svd(scale(df))$v,main="v")
```

```r
z <- rnorm(30, mean = 1, sd=0.5)
ddf <- data.frame(x,y,z)
svd1 <- svd((ddf))
str(svd1)
u <- svd1$u
v <- svd1$v
d <- svd1$d
u%*%diag(d)%*%v - ddf
```
